```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


y = []
z = []
dv = []
t = []
frq = 1000
vpp =1000
```

```python
x = []
with open("/home/csarmiento/MEGA/IyC/Clase 6/entrada_1kHz_2_canales_1000mvpp.txt") as file:
    for l in file.readlines():
        x.append(float(l[:-1] )/41.480)
    
print(len(x) )
```

```python
plt.plot(x[:500])
```

```python
y = list()
for i in range(0,len(x),2):
    y.append(x[i])
print(len(y))
plt.plot(y[:500] )
y = np.array(y)
```

```python
z = list()
for i in range(0, len(x), 2):
    z.append(x[i+1])

z = np.array(z)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.ylabel('Tensión, (mV)', fontsize=16)
plt.plot(z[100:200],'x' )
plt.plot(y[100:200], '.')
plt.grid(True)
plt.savefig('dosCanales.png') 
```

```python
dv = z-y
dt=2000/(frq*vpp)
t = dv*dt   
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.xlabel('tiempo entre canales, (ms)', fontsize=16)
plt.ylabel('Conteo', fontsize=16)
plt.hist(t, histtype="step", bins=50, range=(-5E-3,5E-3), linewidth=2.5, linestyle="-")
plt.grid(True)
plt.savefig('histChannels_1y2.png') 
```
